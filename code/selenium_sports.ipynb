```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
BASE_DIR = os.getcwd()
```

```python
sports_category = ['kbaseball','wbaseball','kfootball','wfootball','basketball','volleyball', 'golf', 'general']
```

```python
from datetime import datetime, timedelta

start_date = datetime(2016, 1, 1)
end_date = datetime(2019, 12, 31)

date_list = []

current_date = start_date
while current_date <= end_date:
    date_list.append(current_date.strftime("%Y%m%d"))
    current_date += timedelta(days=14)
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--incognito")

driver = webdriver.Chrome(options=options)

def get_text(category, date):
    driver.get(f"https://sports.news.naver.com/{category}/news/index?isphoto=N&date={date}&page=1")

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    articles = soup.select('#_newsList li')

    
    titles = []
    
    for article in articles:
        title = article.select_one('.title').text
        titles.append(title)
        
        if len(titles) >= 2:
            break

    
    return titles

total_list = []
for category in sports_category:
    for date in date_list: 
        news_list = get_text(category,date)
        print(news_list)
        total_list.append(news_list)
```

```python
import re
```

```python
df = pd.DataFrame(columns=["text","label"])

for sen_list in total_list:
    for sen in sen_list:
        re_text = re.sub(r'\(|\)|\[|\]|\{|\}|\'|\"|\,|\“|\”|\<|\>', ' ', sen)
        re_text = re.sub(r'\s+', ' ', re_text)
        df.loc[len(df)] = {"text" : re_text, "label":5}
```

```python
for idx, sen in df.iterrows():
    print(sen)
```
