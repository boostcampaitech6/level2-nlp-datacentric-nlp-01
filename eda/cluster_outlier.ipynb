```python
label_array=[[0 for _ in range(7)] for _ in range(7)]

for i in range(7):
    for j in range(7):
        if i==j:
            continue
        label_array[i][j]=len(df[(df['target']==i) & (df['pred']==j)])
```

```python
import matplotlib.pyplot as plt

plt.imshow(label_array, cmap='hot', interpolation='nearest')
plt.show()
```
